```python
import numpy as np
import matplotlib.pyplot as plt

# Caminhos dos arquivos
arquivo_dados = "/home/joaovfg/PFC-WPW/mit-bih-segmented-signals/segments_230.dat"
arquivo_labels = "/home/joaovfg/PFC-WPW/mit-bih-segmented-signals/labels_230.npy"

# Carregar os segmentos
sinal = np.fromfile(arquivo_dados, dtype=np.float32)
segmento_len = 216
num_segmentos = sinal.shape[0] // segmento_len
segmentos = sinal[:num_segmentos * segmento_len].reshape(-1, segmento_len)

# Carregar os r√≥tulos
labels = np.load(arquivo_labels)

# Verificar consist√™ncia
if len(labels) != num_segmentos:
    raise ValueError(f"N√∫mero de labels ({len(labels)}) diferente do n√∫mero de segmentos ({num_segmentos})")

# üî¢ √çndice do segmento a visualizar
indice = 5  # exemplo: segmento 4

# ‚úÖ Verificar e visualizar
if 0 <= indice < num_segmentos:
    plt.figure(figsize=(10, 3))
    plt.plot(segmentos[indice])
    plt.title(f"Segmento {indice + 1} - Classe: {labels[indice]}")
    plt.xlabel("Amostras")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print(f"√çndice inv√°lido. Deve estar entre 0 e {num_segmentos - 1}")
```

```python
# Plotar os primeiros 5 segmentos
for i in range(min(5, num_segmentos)):
    plt.figure(figsize=(10, 3))
    plt.plot(segmentos[i])
    plt.title(f"Segmento {i+1}")
    plt.xlabel("Amostras")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
